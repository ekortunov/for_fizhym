```python
%matplotlib widget
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt

import ipywidgets as ipy
```

## Растворение одной соли (10.2.1)

```python
# set up parameters
zones_number = 3 # number of zones
number_comp = 2 # number of components in both water and rock
c_ini = 0.1 # initial concentrations of mineral in water
c_bound = 0.3 # boundary concentration of minral in water
n_ini = 0.2 #initial concentration of mineral in rock
c_m = 0.4
#c_m = c_bound+n_ini# concentratopn of mineral in equlibrium with boundary solution
v = 0.3#  darcy velocity
n = 0.3 #porosity 
u = np.zeros((1, zones_number-1))
xx = np.zeros((1, zones_number-1))
```

```python
#u = v/n # tranport velocity calculation
# tranport velocitites calculation
u[0,0] = (v/n)/(1+(n_ini/(c_m-c_bound)))# frount 12
u[0,1]  = v/n # front 23
time = np.arange(1,101)
x = np.arange(1,101)
conc1 = np.zeros((number_comp, x.size))# initiate concentration matrix

def transport(t, conc):
        it = np.nditer(u, flags=['f_index'])
        for uu in it:
            xx[0,it.index] = t*uu
       
        conc[0,0:int(np.round(xx[0,0]))] = c_bound
        conc[0,int(np.round(xx[0,0])):int(np.round(xx[0,1]))] = c_m
        conc[0,int(np.round(xx[0,1])):] = c_ini
        #########################################
        conc[1,0:int(np.round(xx[0,0]))] = 0
        conc[1,int(np.round(xx[0,0])):int(np.round(xx[0,1]))] = n_ini
        conc[1,int(np.round(xx[0,1])):] = n_ini
        return conc, xx


#tt = 10
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(1, 1, 1)
transport(0,conc1)
line, = ax.plot(x, transport(0,conc1)[0][0,], label = 'C в воде', linewidth = 3, color = "maroon")
line1, = ax.plot(x, transport(0,conc1)[0][1,],label = 'N в породе', linewidth = 3,color = "tab:blue")
ann = ax.annotate("Выход минерала из породы "+str(xx[0,0]*n_ini), (1, 0.01))
ann1 = ax.annotate("Накопление минерала в воде "+str((xx[0,1]-xx[0,0])*(c_m-c_ini)), (1, 0.05))
out = ipy.Output()
ax.set_xlim([0,conc[1,].size])
ax.set_ylim([0, c_m+0.1])
ax.set_ylabel('C, N, б/р')
ax.set_xlabel('x, б/р')
plt.legend()
def update(time):
    with out:
#        transport(tt,conc1)

        line.set_data(x, transport(slider.value,conc1)[0][0,])
        line1.set_data(x, transport(slider.value,conc1)[0][1,])
        ann.set_text("Выход минерала из породы "+str(np.round(transport(slider.value,conc1)[1][0,0]*n_ini)))
        ann1.set_text("Накопление минерала в воде "+str(np.round((transport(slider.value,conc1)[1][0,1]-transport(slider.value,conc1)[1][0,0])*(c_m-c_bound))))
        fig.canvas.draw_idle()
    

slider = ipy.IntSlider(min = 0, max = 100)
widget = ipy.interactive(update, time = slider)
layout = ipy.Layout(
#     display = 'flex',
#                    flex_flow = 'row',
#                    justify_content = 'space-between',
#                    align_items = 'center',
                   )
widgets = ipy.HBox(children=(slider, out), layout = layout)
display(widgets)
```

## Метасоматическое замещение минералов (11.2.2)

```python
# set up parameters
zones_number = 4 # number of zones
number_comp = 4 # number of components in both water and rock
G_ini = 0.05 # initial concentrations of gypsum in water
G_bound = 0.1 # boundary concentration of gypsum in water
#g_ini = 0.2 #initial concentration of gypsum in rock
G_m = 0.2 #
A_bound = 0.6 # boundary acid concentration
A_m = 0.4 # acid eq conc with calcite
A_ini  = 0.05
q_ini = 0.6  # calcite ini conc
forybound = [q_ini,A_ini,A_m,A_bound,G_m,G_bound,G_ini]
#c_m = c_bound+n_ini# concentratopn of mineral in equlibrium with boundary solution
v = 0.3#  darcy velocity
n = 0.3 #porosity 
u = np.zeros((1, zones_number-1))
xx = np.zeros((1, zones_number-1))
uuu = v/n #transport velocity
```

```python
u[0,0] = uuu/(1+(q_ini/(G_m-G_bound)))
u[0,1] = (uuu)/(1+(q_ini/(A_bound-A_m)))# frount 12
u[0,2]  = uuu  # front 23
time = np.arange(1,101)
x = np.arange(1,101)
conc1 = np.zeros((number_comp, x.size))# initiate concentration matrix

def transport(t, conc):
        it = np.nditer(u, flags=['f_index'])
        for uu in it:
            xx[0,it.index] = t*uu
       ##########################################Кислота
        conc[0,0:int(np.round(xx[0,0]))] = A_bound
        conc[0,int(np.round(xx[0,0])):int(np.round(xx[0,1]))] =  A_bound
        conc[0,int(np.round(xx[0,1])):int(np.round(xx[0,2]))] = A_m
        conc[0,int(np.round(xx[0,2])):] = A_ini
        #########################################Гипс растворенный
        conc[1,0:int(np.round(xx[0,0]))] = G_bound
        conc[1,int(np.round(xx[0,0])):int(np.round(xx[0,1]))] =  G_m
        conc[1,int(np.round(xx[0,1])):int(np.round(xx[0,2]))] = G_m
        conc[1,int(np.round(xx[0,2])):] = G_ini
        ##########################################Кальцит
        conc[2,0:int(np.round(xx[0,0]))] = 0
        conc[2,int(np.round(xx[0,0])):int(np.round(xx[0,1]))] =  0
        conc[2,int(np.round(xx[0,1])):int(np.round(xx[0,2]))] = q_ini
        conc[2,int(np.round(xx[0,2])):] = q_ini
        ######################################### Гипс в пооде
        conc[3,0:int(np.round(xx[0,0]))] = 0
        conc[3,int(np.round(xx[0,0])):int(np.round(xx[0,1]))] =  q_ini
        conc[3,int(np.round(xx[0,1])):int(np.round(xx[0,2]))] =0
        conc[3,int(np.round(xx[0,2])):] = 0
        return conc, xx
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
transport(0,conc1)
line, = ax.plot(x, transport(0,conc1)[0][0,], label = 'Кислота в воде', linewidth = 3.5, color = "maroon")
line1, = ax.plot(x, transport(0,conc1)[0][1,],label = 'Гипс в воде', linewidth = 3,color = "tab:blue")
line2, = ax.plot(x, transport(0,conc1)[0][2,], label = 'Кальцит в породе', linewidth = 3, color = "gold")
line3, = ax.plot(x, transport(0,conc1)[0][3,],'--',label = 'Гипс в породе', linewidth = 3,color = "black")
ann = ax.annotate("Потрачено кислоты из воды "+str(0), (1, -0.07),annotation_clip=False)# неправильно, лень переделывать
ann1 = ax.annotate("Растворено кальцита из породы "+str((0)*(c_m-c_ini)), (1, -0.1),annotation_clip=False,)
ann2 = ax.annotate("Накоплено гипса в воде и породе за счет переотложения "+str((0)*(c_m-c_ini)), (1, -0.13),annotation_clip=False)
out = ipy.Output()
ax.set_xlim([0,conc[1,].size])
ax.set_ylim([0, max(forybound)+0.1])
ax.set_ylabel('C, N, б/р')
ax.set_xlabel('x, б/р')
plt.legend()#loc='center left',bbox_to_anchor=(1,0))
plt.tight_layout()
def update(time):
    with out:
#        transport(tt,conc1)

        line.set_data(x, transport(slider.value,conc1)[0][0,])
        line1.set_data(x, transport(slider.value,conc1)[0][1,])
        line2.set_data(x, transport(slider.value,conc1)[0][2,])
        line3.set_data(x, transport(slider.value,conc1)[0][3,])
        ann.set_text("Потрачено кислоты из воды "+str(np.round((transport(slider.value,conc1)[1][0,2]-transport(slider.value,conc1)[1][0,1])*(A_bound-A_m))))
        ann1.set_text("Растворено кальцита из породы "+str(np.round((transport(slider.value,conc1)[1][0,1])*(q_ini))))
        ann2.set_text("Накоплено гипса в воде и породе за счет переотложения "+str(np.round((transport(slider.value,conc1)[1][0,1]-transport(slider.value,conc1)[1][0,0])*(q_ini)+(G_m-G_bound)*(transport(slider.value,conc1)[1][0,2]-transport(slider.value,conc1)[1][0,0]))))
        fig.canvas.draw_idle()
    

slider = ipy.IntSlider(min = 0, max = 100)
widget = ipy.interactive(update, time = slider)
layout = ipy.Layout(
#     display = 'flex',
#                    flex_flow = 'row',
#                    justify_content = 'space-between',
#                    align_items = 'center',
                   )
widgets = ipy.HBox(children=(slider, out), layout = layout)
display(widgets)
```
